- Read the dataset labelled features.csv.
- Create a ``` lambda ``` function that checks if displacement $>$ 200
- Create a new list that shows true if displacement $>$ 200 and false <br>
if displacement $\leq$ 200.
- Count how many cars have displacement $> 200$.

To get full credit use either the apply or the map function <br>
(see Lectures 7 and 8)

````python
# Write your answer here:
features = pd.read_csv("data/features.csv")

fn_displacement200 = lambda displacement: displacement > 200

features["displacement_over_200"] = features["displacement"].apply(fn_displacement200)

print(features["displacement_over_200"].sum())
````

(b) Subsetting Data

<font size = "3" >

- Read the dataset labelled features.csv.
- Define a global variable called ```threshold``` that takes the <br>
value 2588
- Subset the data to get one dataset with weight $>$ threshold and <br>
another with weight $\le$ 2588. Use the ```.query()``` function.
- Compute the mean acceleration for each of these datasets.<br>

Note: Review how to define a global variable inside ```.query()```
 

````python
# Write your answer here:
features = pd.read_csv("data/features.csv")

threshold = 2588
features_heavy = features.query("weight > @threshold")
features_light = features.query("weight <= @threshold")

print(features_heavy["acceleration"].mean())
print(features_light["acceleration"].mean())
````

(e) Plot by Category

<font size = "3">

Read the "data/wdi_urban.csv" dataset.

  This contains data from 1980 and 2020.

- **Create a list with the unique values of "year"**

- **Create a plot with a ...**
    - Histogram of "prop_urbanpopulation" for each year
    - Set the histogram option `alpha = 0.5` for transparency
    - All the histograms should be overlapping
    - Create a legend with the list of years
    - Label the axes and the title


````python
# Write your answer here:
urban = pd.read_csv("data/wdi_urban.csv")
urban

unique_years = pd.unique(urban["year"])

for year in unique_years:
    df = urban.query("year == @year")
    plt.hist(x = df["prop_urbanpopulation"], alpha=0.5, label=str(year)) 

plt.xlabel("Proportion of Urban Population")
plt.ylabel("Frequency")
plt.title("Histogram of Urban Population Proportion (1980 vs 2020)")
plt.legend(labels = unique_years,
           title="Year")
plt.show()
````
